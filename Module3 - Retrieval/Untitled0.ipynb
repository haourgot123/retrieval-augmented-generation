## 1.4 Document Transformers

```python
"""Sử dụng RecursiveCharacterTextSplitter để split document đã load phía trên"""

from langchain.text_splitter import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=50,
    length_function=len,
    add_start_index=True,
)

text_recursive = text_splitter.split_documents(document)
print(text_recursive[0].page_content)
print("=" * 50)
print(text_recursive[1].page_content)
```

```python
"""Sử dụng CharacterTextSplitter để split document đã load phía trên"""


from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    length_function=len,
    add_start_index=True,
)

text_chracter = text_splitter.split_documents(document)
print(text_chracter[0].page_content)
print("=" * 50)
print(text_chracter[1].page_content)
```

```python
"""LangChain cung cấp các hàm chia cụ thể cho các ngôn ngữ lập trình khác nhau, như bộ chia mã Python và bộ chia mã JavaScript:"""

from langchain.text_splitter import RecursiveCharacterTextSplitter, Language

python_code = """
def hello_world():
    print("Hello, World!")
hello_world()
"""

python_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.PYTHON, chunk_size=200
)
python_docs = python_splitter.create_documents([python_code])
print(python_docs[0])

js_code = """
function helloWorld() {
  console.log("Hello, World!");
}
helloWorld();
"""

js_splitter = RecursiveCharacterTextSplitter.from_language(
    language=Language.JS, chunk_size=200
)
js_docs = js_splitter.create_documents([js_code])
print(js_docs[0])
```

## 2.TEXT EMBEDDING MODELS

```python
from langchain_community.embeddings.fastembed import FastEmbedEmbeddings


embedding_model = FastEmbedEmbeddings(model_name="BAAI/bge-small-en-v1.5")


embeddings = embedding_model.embed_documents(
    ["Xin chào",
     "Thành phố Hưng Yên đổi mới với nhiều phương pháp cải cách mới",
     "Hà Nội được đánh giá là thành phố có giá cả đăt đỏ nhât cả nước",
     "Việt Nam là đất nước được nhiều du khách nước ngoài yêu thích"]
)

print("Number of documents embedded:", len(embeddings))
print("Dimension of each embedding:", len(embeddings[0]))
```

# 3. VECTOR STORES

## 3.1 Vector embedding with Chroma

```python
from langchain.vectorstores import Chroma

chroma = Chroma.from_documents(text_recursive, embedding_model)

similar_texts = chroma.similarity_search("Liệt kê các sản phẩm bếp từ")
print(similar_texts[0].page_content)

print("=" * 50)

print(similar_texts[1].page_content)
```

## 3.2 Vector embedding with FAISS

```python
from langchain.vectorstores import FAISS

faiss = FAISS.from_documents(text_recursive, embedding_model)
similar_faiss = faiss.similarity_search("Số lượng các sản phẩm bếp từ")
print(similar_faiss[0].page_content)
```

```python
print(similar_faiss[1].page_content)
```

```python
print(similar_faiss[2].page_content)
```

## 3.3 Vector embedding with QDrant

### 3.3.1 Search similarity

```python
from langchain.vectorstores import Qdrant

qdrant = Qdrant.from_documents(
    text_recursive,
    embedding_model,
    location=":memory:",  # Local mode with in-memory storage only
)
similar_qdrant = qdrant.similarity_search("có bao nhiêu sản phẩm bếp từ ?")
print(similar_qdrant[0].page_content)
```

### 3.3.2 Search similarity search with score

```python
from langchain.vectorstores import Qdrant

"""Similarity search with score"""

qdrant = Qdrant.from_documents(
    text_recursive,
    embedding_model,
    location=":memory:",  # Local mode with in-memory storage only
)
similar_qdrant = qdrant.similarity_search_with_score("có bao nhiêu sản phẩm bếp từ ?")
print("Query similar: \n", similar_qdrant[0][0].page_content)
print("Score: \n", similar_qdrant[0][1])
```

```python
print("Query similar: \n", similar_qdrant[1][0].page_content)
print("Score: \n", similar_qdrant[1][1])
```

### 3.3.3 Maximum marginal relevance search (MMR)

```python
query = "Liệt kê các sản phẩm điều hòa"

found_docs = qdrant.max_marginal_relevance_search(query, k=2, fetch_k=10)

for i, doc in enumerate(found_docs):
    print(f"{i + 1}.", doc.page_content, "\n")
```

# 4. RETRIEVERS

## 4.1 Simple Retriever

```python
"""Retriever từ vector db của faiss đã được tạo bên trên"""

retriever = faiss.as_retriever()

query = "sản phảm nồi cơm điện 509172"
retrieved_docs = retriever.invoke(query)
print(retrieved_docs[0].page_content)
```

## 4.2 MultiQueryRetriever

```python
from langchain_groq import ChatGroq
from langchain.retrievers.multi_query import MultiQueryRetriever
from google.colab import userdata


"""Sử dụng MultiQueryRetriever để truy vấn"""

query = "Liệt kê sản phảm nồi cơm điện"

llm = ChatGroq(model="llama3-70b-8192",
                api_key= userdata.get("GROQ_API_KEY"),
                temperature=0.1)


retriever_from_llm = MultiQueryRetriever.from_llm(
    retriever=faiss.as_retriever(), llm=llm
)
unique_docs = retriever_from_llm.get_relevant_documents(query=query)
print("Number of unique documents:", len(unique_docs))
print(unique_docs[0].page_content)
```

# 4.3 Contextual Compression

```python
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor

"""Sử dụng Contextual Compression để truy vấn"""


llm = ChatGroq(model="llama3-70b-8192",
                api_key=userdata.get("GROQ_API_KEY"),
                temperature=0.1)

compressor = LLMChainExtractor.from_llm(llm)
compression_retriever = ContextualCompressionRetriever(base_compressor=compressor, base_retriever=retriever)

compressed_docs = compression_retriever.get_relevant_documents(query)
print(compressed_docs[0].page_content)
```
