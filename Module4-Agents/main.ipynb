## 1.1 DUCKDUCKGO

```python
from langchain.tools import DuckDuckGoSearchRun
search = DuckDuckGoSearchRun()
search.run("Thủ đô của nước Việt Nam là gì ?")
```

```python
from langchain_community.tools.tavily_search import TavilySearchResults
from google.colab import userdata
import os
os.environ["TAVILY_API_KEY"] = userdata.get('TAVILY_API_KEY')
search = TavilySearchResults()

tools = [search]

results = search.invoke({ # tự top_k
    "query": "ưu và nhược điểm của Điều hòa Daikin Inverter",
    "include_answer": False,   # True: có thể tự trả lời query
})

for result in results:
    print(result['url'])
    print(result['content'])
    print("-" * 100)
```

# 2.Agents

```python
from langchain_groq import ChatGroq



llm = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)


llm.invoke('Có bao nhiêu kí tự trong từ VIETNAM')
```

## 2.1  Custom Tools

```python
from langchain.agents import tool

@tool
def get_word_len(word: str) -> int:
    """Returns the length of a word."""
    return len(word)

get_word_len.invoke('VietNam')
```

```python
from langchain.prompts import ChatPromptTemplate, MessagesPlaceholder


prompts = ChatPromptTemplate.from_messages([
    ("system", "Bạn là một trợ lý ảo nhưng không giỏi trong việc tính toán độ dài chữ"),
    ("user", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])
```

```python
from langchain.agents.format_scratchpad.openai_tools import format_to_openai_tool_messages
from langchain.agents.output_parsers.openai_tools import OpenAIToolsAgentOutputParser
llm_with_tools = llm.bind_tools(tools)

agent = (
    {
        'input': lambda x: x['input'],
        'agent_scratchpad': lambda x: format_to_openai_tool_messages(x['intermediate_steps'])
    }
    | prompts
    | llm_with_tools
    | OpenAIToolsAgentOutputParser()

)
```

```python
agent
```

```python
from langchain.agents import AgentExecutor

agent_executor = AgentExecutor(agent=agent, tools=tools, verbose=True)
output = agent_executor.stream({"input": "How many letters in the word NGUYENNGUYENHAO"})
# print(list(output))
print(list(output)[-1]['output'])
```
