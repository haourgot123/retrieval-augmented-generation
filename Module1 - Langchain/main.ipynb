# 1. Káº¿t há»£p Pompt + LLM

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from google.colab import userdata


prompt = ChatPromptTemplate.from_template(
    "HÃ£y trÃ¬nh bÃ y vá» Ä‘áº¥t nÆ°á»›c {foo}. LÆ°u Ã½ chá»‰ sá»­ dá»¥ng Tiáº¿ng Viá»‡t Ä‘á»ƒ trá»Ÿ lá»i"
)


model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

chain = prompt | model
response = chain.invoke({"foo": "Viá»‡t Nam"})
print(response.content)
```

# 3.RAG (Retrieval Augmented Generation)

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain.embeddings import FastEmbedEmbeddings
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough, RunnableLambda
from langchain.vectorstores import FAISS

embedding_model = FastEmbedEmbeddings(model_name="BAAI/bge-small-en-v1.5")
# Táº¡o vectorstore lÆ°u 3 cÃ¢u Ä‘á»ƒ retrieval
vectorstore = FAISS.from_texts(
    ["NÆ°á»›c Nga, tÃªn Ä‘áº§y Ä‘á»§ lÃ  LiÃªn bang Nga, lÃ  má»™t quá»‘c gia náº±m táº¡i Báº¯c Ã‚u vÃ  ÄÃ´ng Ã‚u.",
     " ÄÃ¢y lÃ  quá»‘c gia lá»›n nháº¥t tháº¿ giá»›i vá» diá»‡n tÃ­ch, vá»›i tá»•ng diá»‡n tÃ­ch khoáº£ng 17,1 triá»‡u kmÂ², chiáº¿m khoáº£ng 1/8 diá»‡n tÃ­ch Ä‘áº¥t liá»n cá»§a TrÃ¡i Äáº¥t.",
     "Thá»§ Ä‘Ã´ cá»§a Nga lÃ  Moscow, thÃ nh phá»‘ lá»›n nháº¥t vÃ  lÃ  trung tÃ¢m kinh táº¿, chÃ­nh trá»‹, vÄƒn hÃ³a cá»§a nÆ°á»›c nÃ y."], embedding=embedding_model
)
retriever = vectorstore.as_retriever()

# Äá»‹nh nghÄ©a 1 template cho prompting
template = ''' Tráº£ lá»i cÃ¢u há»i dá»±a vÃ o thÃ´ng tin dÆ°á»›i Ä‘Ã¢y:
{context}
LÆ°u Ã½: Báº¡n chá»‰ Ä‘Æ°á»£c tráº£ lá»i báº±ng tiáº¿ng viá»‡t
Question: {question}
'''
# ÄÆ°a Template vÃ o LangChain
prompt = ChatPromptTemplate.from_template(template)

model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

# Táº¡o RAG

chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
response = chain.invoke("HÃ£y trÃ¬nh bÃ y vá» Ä‘áº¥t nÆ°á»›c Nga")
print(response)
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough, RunnableLambda
from google.colab import userdata
from operator import itemgetter
model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

planner = (
    ChatPromptTemplate.from_template('Táº¡o láº­p luáº­n vá» {input}. LÆ°u Ã½: Chá»‰ tráº£ lá»i báº±ng tiáº¿ng viá»‡t')
    | model
    | { 'base_response' : StrOutputParser()}
)

arguments_for = (
    ChatPromptTemplate.from_template(
        'LiÃªt kÃª Æ°u Ä‘iá»ƒm hoáº·c nhá»¯ng khÃ­a cáº¡nh tÃ­ch cá»±c cá»§a {base_response}. LÆ°u Ã½: Báº¡n chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng TiÃªng Viá»‡t'
    )
    | model
    | StrOutputParser()
)

arguments_against = (
    ChatPromptTemplate.from_template(
        'Liá»‡t kÃª nhÆ°á»£c Ä‘iá»ƒm hoáº·c nhá»¯ng khÃ­a cáº¡nh tiÃªu cá»±c cá»§a {base_response}. LÆ°u Ã½: Báº¡n chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng TiÃªng Viá»‡t'
    )
    | model
    | StrOutputParser()
)

final_responder = (
    ChatPromptTemplate.from_messages(
        [
            ("ai", "{original_response}"),
            ("human", "Pros:\n{results_1}\n\nCons:\n{results_2}"),
            ("system", "Táº¡o pháº£n há»“i cuá»‘i cÃ¹ng dá»±a trÃªn lá»i phÃª bÃ¬nh. LÆ°u Ã½ chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng Tiáº¿ng Viá»‡t"),
        ]
    )
    | model
    | StrOutputParser()
)

chain = (
    planner
    | {
        "results_1": arguments_for,
        "results_2": arguments_against,
        "original_response": itemgetter("base_response"),
    }
    | final_responder
)

result = chain.invoke({"input": "mÃ´ hÃ¬nh ngÃ´n ngá»¯ lá»›n"})
print(result)
```

# Adding Memory to a Chain

```python
from langchain.schema import SystemMessage
from langchain.prompts import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    MessagesPlaceholder,
)
from langchain.memory import ConversationBufferMemory
from langchain_groq import ChatGroq
from langchain.chains import LLMChain



prompt = ChatPromptTemplate.from_messages(
    [
      SystemMessage(
          content = 'Báº¡n lÃ  má»™t chatbot há»¯u Ã­ch Ä‘ang trÃ² chuyá»‡n vá»›i con ngÆ°á»i. LÆ°u Ã½: Báº¡n chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng Tiáº¿ng Viá»‡t'
      ),
      MessagesPlaceholder(
          variable_name="chat_history"
      ), # NÆ¡i lÆ°u trá»¯ lá»‹ch sá»­ Ä‘oáº¡n chat
      HumanMessagePromptTemplate.from_template(
          "{human_input}"
      ),
    ]
)

llm = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)

chain = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=memory
)



response = chain.predict(human_input="Viá»‡t Nam cÃ³ bao nhiÃªu dÃ¢n tá»™c?")
print(response)
```

# 7. Routing by Semantic Similarity

```python
from langchain.chat_models import ChatOpenAI
from langchain.embeddings import OpenAIEmbeddings
from langchain.prompts import PromptTemplate
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnableLambda, RunnablePassthrough
from langchain.utils.math import cosine_similarity
from langchain.embeddings import FastEmbedEmbeddings
physics_template = """Báº¡n lÃ  má»™t giÃ¡o sÆ° váº­t lÃ½ ráº¥t thÃ´ng minh. \
Báº¡n ráº¥t giá»i tráº£ lá»i cÃ¡c cÃ¢u há»i vá» váº­t lÃ½ má»™t cÃ¡ch ngáº¯n gá»n vÃ  dá»… hiá»ƒu. \
Khi báº¡n khÃ´ng biáº¿t cÃ¢u tráº£ lá»i cho má»™t cÃ¢u há»i, báº¡n thá»«a nháº­n ráº±ng báº¡n khÃ´ng biáº¿t
LÆ°u Ã½: báº¡n chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng tiáº¿ng viá»‡t Ä‘á»ƒ trÃ² chuyá»‡n
Here is a question:
{query}"""

math_template = """Báº¡n lÃ  má»™t nhÃ  toÃ¡n há»c ráº¥t giá»i. Báº¡n ráº¥t giá»i tráº£ lá»i cÃ¡c cÃ¢u há»i toÃ¡n há»c. \
Báº¡n tháº­t giá»i vÃ¬ báº¡n cÃ³ thá»ƒ chia nhá» cÃ¡c váº¥n Ä‘á» khÃ³ khÄƒn thÃ nh cÃ¡c pháº§n cáº¥u thÃ nh cá»§a chÃºng, \
tráº£ lá»i cÃ¡c pháº§n cáº¥u thÃ nh vÃ  sau Ä‘Ã³ ghÃ©p chÃºng láº¡i vá»›i nhau Ä‘á»ƒ tráº£ lá»i cÃ¢u há»i rá»™ng hÆ¡n.
LÆ°u Ã½: báº¡n chá»‰ Ä‘Æ°á»£c sá»­ dá»¥ng tiáº¿ng viá»‡t Ä‘á»ƒ trÃ² chuyá»‡n
Here is a question:
{query}"""

embedding_model = FastEmbedEmbeddings(model_name="BAAI/bge-small-en-v1.5")

prompt_templates = [physics_template, math_template]
prompt_embeddings = embedding_model.embed_documents(prompt_templates)
print(len(prompt_embeddings))

def prompt_router(input):
    query_embedding = embedding_model.embed_query(input["query"])
    similarity = cosine_similarity([query_embedding], prompt_embeddings)[0]
    most_similar = prompt_templates[similarity.argmax()]
    print("Using MATH" if most_similar == math_template else "Using PROGRAM")
    return PromptTemplate.from_template(most_similar)

chain = (
    {"query": RunnablePassthrough()}
    | RunnableLambda(prompt_router)
    | llm
    | StrOutputParser()
)

print(chain.invoke({"query": "Lá»— Ä‘en lÃ  gÃ¬?"}))
print(chain.invoke({"query": "NgÃ´n ngá»¯ láº­p trÃ¬nh Java Ä‘á»ƒ lÃ m gÃ¬?"}))
```
