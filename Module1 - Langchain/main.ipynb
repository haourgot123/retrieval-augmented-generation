# 1. Kết hợp Pompt + LLM

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from google.colab import userdata


prompt = ChatPromptTemplate.from_template(
    "Hãy trình bày về đất nước {foo}. Lưu ý chỉ sử dụng Tiếng Việt để trở lời"
)


model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

chain = prompt | model
response = chain.invoke({"foo": "Việt Nam"})
print(response.content)
```

# 2. Prompt  + LLM + OutputParser

🔥 Kết hợp OutputParser để cấu trúc lại đầu ra từ LLM

```python
from langchain.schema.output_parser import StrOutputParser

chain = prompt | model | StrOutputParser()
response = chain.invoke({"foo": "Việt Nam"})
print(response)
```

# 3.RAG (Retrieval Augmented Generation)

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain.embeddings import FastEmbedEmbeddings
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough, RunnableLambda
from langchain.vectorstores import FAISS

embedding_model = FastEmbedEmbeddings(model_name="BAAI/bge-small-en-v1.5")
# Tạo vectorstore lưu 3 câu để retrieval
vectorstore = FAISS.from_texts(
    ["Nước Nga, tên đầy đủ là Liên bang Nga, là một quốc gia nằm tại Bắc Âu và Đông Âu.",
     " Đây là quốc gia lớn nhất thế giới về diện tích, với tổng diện tích khoảng 17,1 triệu km², chiếm khoảng 1/8 diện tích đất liền của Trái Đất.",
     "Thủ đô của Nga là Moscow, thành phố lớn nhất và là trung tâm kinh tế, chính trị, văn hóa của nước này."], embedding=embedding_model
)
retriever = vectorstore.as_retriever()

# Định nghĩa 1 template cho prompting
template = ''' Trả lời câu hỏi dựa vào thông tin dưới đây:
{context}
Lưu ý: Bạn chỉ được trả lời bằng tiếng việt
Question: {question}
'''
# Đưa Template vào LangChain
prompt = ChatPromptTemplate.from_template(template)

model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

# Tạo RAG

chain = (
    {"context": retriever, "question": RunnablePassthrough()}
    | prompt
    | model
    | StrOutputParser()
)
response = chain.invoke("Hãy trình bày về đất nước Nga")
print(response)
```

```python
from langchain.prompts import ChatPromptTemplate
from langchain_groq import ChatGroq
from langchain.schema.output_parser import StrOutputParser
from langchain.schema.runnable import RunnablePassthrough, RunnableLambda
from google.colab import userdata
from operator import itemgetter
model = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

planner = (
    ChatPromptTemplate.from_template('Tạo lập luận về {input}. Lưu ý: Chỉ trả lời bằng tiếng việt')
    | model
    | { 'base_response' : StrOutputParser()}
)

arguments_for = (
    ChatPromptTemplate.from_template(
        'Liêt kê ưu điểm hoặc những khía cạnh tích cực của {base_response}. Lưu ý: Bạn chỉ được sử dụng Tiêng Việt'
    )
    | model
    | StrOutputParser()
)

arguments_against = (
    ChatPromptTemplate.from_template(
        'Liệt kê nhược điểm hoặc những khía cạnh tiêu cực của {base_response}. Lưu ý: Bạn chỉ được sử dụng Tiêng Việt'
    )
    | model
    | StrOutputParser()
)

final_responder = (
    ChatPromptTemplate.from_messages(
        [
            ("ai", "{original_response}"),
            ("human", "Pros:\n{results_1}\n\nCons:\n{results_2}"),
            ("system", "Tạo phản hồi cuối cùng dựa trên lời phê bình. Lưu ý chỉ được sử dụng Tiếng Việt"),
        ]
    )
    | model
    | StrOutputParser()
)

chain = (
    planner
    | {
        "results_1": arguments_for,
        "results_2": arguments_against,
        "original_response": itemgetter("base_response"),
    }
    | final_responder
)

result = chain.invoke({"input": "mô hình ngôn ngữ lớn"})
print(result)
```

# Adding Memory to a Chain

```python
from langchain.schema import SystemMessage
from langchain.prompts import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    MessagesPlaceholder,
)
from langchain.memory import ConversationBufferMemory
from langchain_groq import ChatGroq
from langchain.chains import LLMChain



prompt = ChatPromptTemplate.from_messages(
    [
      SystemMessage(
          content = 'Bạn là một chatbot hữu ích đang trò chuyện với con người. Lưu ý: Bạn chỉ được sử dụng Tiếng Việt'
      ),
      MessagesPlaceholder(
          variable_name="chat_history"
      ), # Nơi lưu trữ lịch sử đoạn chat
      HumanMessagePromptTemplate.from_template(
          "{human_input}"
      ),
    ]
)

llm = ChatGroq(
    model = "llama3-70b-8192",
    api_key = userdata.get('GROQ_API_KEY')
)

memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)

chain = LLMChain(
    llm=llm,
    prompt=prompt,
    verbose=True,
    memory=memory
)



response = chain.predict(human_input="Việt Nam có bao nhiêu dân tộc?")
print(response)
```

# 7. Routing by Semantic Similarity
